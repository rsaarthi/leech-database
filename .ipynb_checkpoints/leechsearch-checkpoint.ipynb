```python
#THis code works with the column name - need to specify with which column you need to perform the search function from
def search_dataframe(df):
    # Get the column name from the user
    column_name = input("Enter the column name to search in: ")

    # Check if the column exists in the DataFrame
    if column_name not in df.columns:
        print(f"Column '{column_name}' does not exist in the DataFrame.")
        return
    
    # Get the search value from the user
    search_value = input(f"Enter the value to search for in column '{column_name}': ")
    
    # Search for the value in the specified column
    matching_rows = df[df[column_name] == search_value]

    # Check if any matching rows were found
    if not matching_rows.empty:
        print(f"Found the following rows where {column_name} is '{search_value}':")
        print(matching_rows)
    else:
        print(f"No rows found where {column_name} is '{search_value}'.")

# Call the search function
search_dataframe(df)
```

```python
# this works for general search, there is no need to specify which column you need to perfrom the search function 
def search_dataframe(df):
    # Get the search value from the user
    search_value = input("Enter the value to search for: ")

    # Initialize an empty DataFrame to store matching rows
    matching_rows = pd.DataFrame()

    # Iterate through each column and search for the value
    for column in df.columns:
        matching_rows = pd.concat([matching_rows, df[df[column] == search_value]])

    # Drop duplicate rows that might occur if the search value matches in multiple columns
    matching_rows = matching_rows.drop_duplicates()

    # Check if any matching rows were found
    if not matching_rows.empty:
        print(f"Found the following rows where any column matches '{search_value}':")
        print(matching_rows)
    else:
        print(f"No rows found where any column matches '{search_value}'.")

# Call the search function
search_dataframe(df)
```
