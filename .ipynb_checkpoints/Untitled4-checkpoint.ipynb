```python
import pandas as pd

# Replace 'leech_data.csv' with the path to your CSV file
df = pd.read_csv('leech_data.csv')

# Get the basic information of the table
df.info()

# Get the count of each species name
counts = df['Species name'].value_counts()
print(counts)

# This works for a general search, with no need to specify which column to search in
def search_dataframe(df):
    # Get the search value from the user
    search_value = input("Enter the value to search for: ").strip().lower()

    # Initialize an empty DataFrame to store matching rows
    matching_rows = pd.DataFrame()

    # Iterate through each column and search for the value
    for column in df.columns:
        matching_rows = pd.concat([matching_rows, df[df[column].astype(str).str.lower() == search_value]], 
                                  ignore_index=True)

    # Drop duplicate rows that might occur if the search value matches in multiple columns
    matching_rows = matching_rows.drop_duplicates()

    # Check if any matching rows were found
    if not matching_rows.empty:
        print(f"Found the following rows where any column matches '{search_value}':")
        print(matching_rows)
    else:
        print(f"No rows found where any column matches '{search_value}'.")

# Call the search function
search_dataframe(df)
```

```python
import pandas as pd

# Replace 'leech_data.csv' with the path to your CSV file
df = pd.read_csv('leech_data.csv')

# Get the basic information of the table
df.info()

# Get the count of each species name
counts = df['Species name'].value_counts()
print(counts)

# This function searches the DataFrame for partial matches across all columns
def search_dataframe(df):
    # Get the search value from the user
    search_value = input("Enter the value to search for: ").strip().lower()

    # Initialize an empty DataFrame to store matching rows
    matching_rows = pd.DataFrame()

    # Iterate through each column and search for the value
    for column in df.columns:
        matching_rows = pd.concat([matching_rows, df[df[column].astype(str).str.lower().str.contains(search_value)]], 
                                  ignore_index=True)

    # Drop duplicate rows that might occur if the search value matches in multiple columns
    matching_rows = matching_rows.drop_duplicates()

    # Check if any matching rows were found
    if not matching_rows.empty:
        print(f"Found the following rows where any column contains '{search_value}':")
        print(matching_rows)
    else:
        print(f"No rows found where any column contains '{search_value}'.")

# Call the search function
search_dataframe(df)
```
